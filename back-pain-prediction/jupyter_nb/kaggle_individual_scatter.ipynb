```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/original.csv')
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

x = data.degree_spondylolisthesis
y = data.pelvic_incidence

plt.xlabel('degree spondylolisthesis')
plt.ylabel('pelvic incidence')

ax.scatter(x, y, color=['b','g'])
plt.show()
```
