```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import matplotlib as mpl
import seaborn as sns
dpi=200
mpl.rcParams['figure.dpi']= dpi
```

```python
df = pd.read_csv('../data/original.csv')
```

```python
for status in ['Normal', 'Abnormal']:
    pi_status = df.pelvic_incidence[df.Class_att == status]
    
    # Draw the density plot
    sns.distplot(pi_status, hist = False, kde = True,
                 kde_kws = {'linewidth': 3},
                 label = status)

```
